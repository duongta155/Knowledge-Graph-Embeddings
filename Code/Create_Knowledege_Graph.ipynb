## Convert all industry numbers to industry names

```python
def convert_industry_number(indus_num_lis):
  indus_name_lis = []
  if indus_num_lis == '-1':
    return indus_num_lis
  for i in indus_num_lis:
    if i != '-1':
      indus_name_lis.append(indus_num_to_indus_name[i])
  return indus_name_lis
```

```python
df_mongo_updated['main_industry'] = df_mongo_updated['main_industry'].apply(lambda x: indus_num_to_indus_name[x] if x != '-1' else x)
```

```python
df_mongo_updated['industries'] = df_mongo_updated['industries'].map(convert_industry_number)
```

### Generate all features for every company

```python
len(list(df_mongo_updated['name'].unique()))
```

```python
start = time.time()

for index, r in tqdm(df_mongo_updated.iterrows()):
    #Extract Variable
    t_com = r['name']
    t_ind = r['industries']
    t_mainind = r['main_industry']
    t_SE = r['scale.equity']
    t_SP = r['scale.profit']
    t_SA = r['scale.asset']
    t_SP = r['scale.profit']
    t_SR = r['scale.revenue']
    t_tax = r['tax_code']
    t_lat = r['cate_latitude']
    t_long = r['bin_longitude']
    
    #Fill the knowledge graph
    tax_com_dict[t_tax] = t_com
    com_ent_lis.append(t_com)
    for i in t_ind:
        if i != '-1':
            know_lis.append([t_com, "Have_Industry", i])
            #if i not in indus_ent_lis:
                #indus_ent_lis.append(i)
            indus_ent_lis.add(i)
    if (t_mainind != '-1'):
        know_lis.append([t_com, "Have_Main_Industry", t_mainind])
        #if t_mainind not in indus_ent_lis:
            #indus_ent_lis.append(t_mainind)
        indus_ent_lis.add(t_mainind)
    if (t_SE != '-_SE'):
        know_lis.append([t_com, "Have_Scale_Equity", t_SE])
        #if t_SE not in ent_lis:
            #ent_lis.append(t_SE)
    if (t_SP != '-_SP'):
        know_lis.append([t_com, "Have_Scale_Profit", t_SP])
        #if t_SP not in ent_lis:
            #ent_lis.append(t_SP)
    if (t_SR != '-_SR'):
        know_lis.append([t_com, "Have_Scale_Revenue", t_SR])
        #if t_SR not in ent_lis:
            #ent_lis.append(t_SR)
    if (t_SA != '-_SA'):
        know_lis.append([t_com, "Have_Scale_Asset", t_SA])
        #if t_SA not in ent_lis:
            #ent_lis.append(t_SA)
    if (t_lat != '-1'):
        know_lis.append([t_com, "Have_Latitude", t_lat])
        #if t_lat not in ent_lis:
            #ent_lis.append(t_lat)
    if (t_long != '-1'):
        know_lis.append([t_com, "Have_Longitude", t_long])
        #if t_long not in ent_lis:
            #ent_lis.append(t_long)        

end = time.time()
print("Time taken to create knowledge graph of companies: ", (end - start))
```
