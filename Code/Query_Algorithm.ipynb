```python
def eval_scale(q_SA, q_SE, q_SP, q_SR, t_SA, t_SE, t_SP, t_SR):
  if (q_SA == '-' and q_SE == '-' and q_SP == '-' and q_SR == '-'):
        return -1
  
  score_SA = -1
  score_SE = -1
  score_SP = -1
  score_SR = -1

  if (t_SA != '-' and q_SA != '-'):
    score_SA = (10 - abs(ord(q_SA) - ord(t_SA))) * 1.0 / 10
  elif ((t_SA != '-' and q_SA == '-') or (q_SA != '-' and t_SA == '-')):
    score_SA = 0

  if (t_SE != '-' and q_SE != '-'):
    score_SE = (10 - abs(ord(q_SE) - ord(t_SE))) * 1.0 / 10
  elif ((t_SE != '-' and q_SE == '-') or (q_SE != '-' and t_SE == '-')):
    score_SE = 0

  if (t_SP != '-' and q_SP != '-'):
    score_SP = (10 - abs(ord(q_SP) - ord(t_SP))) * 1.0 / 10
  elif ((t_SP != '-' and q_SP == '-') or (q_SP != '-' and t_SP == '-')):
    score_SP = 0

  if (t_SR != '-' and q_SR != '-'):
    score_SR = (10 - abs(ord(q_SR) - ord(t_SR))) * 1.0 / 10
  elif ((t_SR != '-' and q_SR == '-') or (q_SR != '-' and t_SR == '-')):
    score_SR = 0
  
  fin_score = [score_SA, score_SE, score_SP, score_SR]
  while (-1 in fin_score):
    fin_score.remove(-1)
    
  if (len(fin_score) == 0):
    return -1
  return np.average(fin_score)
```

```python
def eval_location(q_lat, q_long, t_lat, t_long):
  score_lat = -1
  score_long = -1

  if t_lat != '-1' and q_lat != '-1':
    score_lat = (n_lat - abs(lis_lat.index(t_lat) - lis_lat.index(q_lat))) * 1.0 / n_lat
  elif ((t_lat != '-1' and q_lat == '-1') or (t_lat == '-1' and q_lat != '-1')):
    score_lat = 0

  if t_long != '-1' and q_long != '-1':
    score_long = (n_long - abs(lis_long.index(t_long) - lis_long.index(q_long))) * 1.0 / n_long
  elif ((t_long != '-1' and q_long == '-1') or (t_long == '-1' and q_long != '-1')):
    score_long = 0
  
  fin_loc = [score_lat, score_long]
  while (-1 in fin_loc):
    fin_loc.remove(-1)
  
  if (len(fin_loc) == 0):
    return -1
  return np.average(fin_loc)
```

```python
def eval_main_industry(q_mainindus, t_mainindus):
  if (q_mainindus == '-1' and t_mainindus == '-1'):
    return -1

  if ((q_mainindus != '-1' and t_mainindus  == '-1') or (q_mainindus == '-1' and t_mainindus != '-1')):
    return 0

  if q_mainindus == t_mainindus:
    return 1

  q_mi_index = indus_name_index_dict[q_mainindus]
  t_mi_index = indus_name_index_dict[t_mainindus]
  temp_dist = industry_distance[q_mi_index][t_mi_index]
  if temp_dist == 100000:
    return 0
  else:
    return (n_distance - temp_dist + 1) / (n_distance + 1)
```

```python
def gen_industry_by_level(industries, level):
  res = set()
  if level == 0:
    return set(industries)
  else:
    for i in industries:
      for j in indus_tree[i]['parent']:
        upper_industry = j
        res.add(upper_industry)
    return res
```

```python
def overlap_similarity(set_1, set_2):
  if max(len(set_1), len(set_2)) == 0:
    return -1
  return len(set_1.intersection(set_2))/max(len(set_1), len(set_2))


def eval_industries(query_industries, result_industries):
  res = []
  # for level in range(3):
  #   _query_industries = gen_industry_by_level(query_industries, level)
  #   _result_industries = gen_industry_by_level(result_industries, level)
  #   score = overlap_similarity(_query_industries, _result_industries)
  #   res.append(score)

  _query_industries = gen_industry_by_level(query_industries, 0)
  _result_industries = gen_industry_by_level(result_industries, 0)
  score = overlap_similarity(_query_industries, _result_industries)
  res.append(score)

  while (-1 in res):
    res.remove(-1)
  if (len(res) == 0):
    return -1
  return np.average(res)

    
```

```python
def eval_single_similarity(q_com, t_com):
  q_arr = com_fea_dict[q_com]
  t_arr = com_fea_dict[t_com]
  similarity_result = [eval_scale(q_arr[5], q_arr[6], q_arr[7], q_arr[8], t_arr[5], t_arr[6], t_arr[7], t_arr[8]),
                       eval_location(q_arr[12], q_arr[13], t_arr[12], t_arr[13]),
                       eval_main_industry(q_arr[4], t_arr[4]),
                       eval_industries(q_arr[3], t_arr[3])]
  while (-1 in similarity_result):
    similarity_result.remove(-1)

  if (len(similarity_result) == 0):
    return -1
  return np.average(similarity_result)
```

### Define total metric

```python
def eval_total_similarity(q_com, t_comlist):
  similarity_result = []
  for i in t_comlist:
    similarity_result.append(eval_single_similarity(q_com, i))
  
  while (-1 in similarity_result):
    similarity_result.remove(-1)
  
  if (len(similarity_result) == 0):
    return -1
  return np.average(similarity_result)
```

# Extract tax_code to find similar companies

## Preprocess required data for extraction process

### Read tax code from pickle

```python
tax_com_dict = pickle.load(open("./Data/Dataset_Mongo_CTEL/Tax_Company_Lookup_Dictionary.pkl","rb"))
```

### Create a company - index dictionary

```python
com_index_dictionary = defaultdict()
```

```python
for i in range(len(ent_lis)):
  com_index_dictionary[ent_lis[i]] = i
```

### Create a company - tax code dictionary

```python
com_tax_dict = {v:k for (k,v) in tax_com_dict.items()}
```

## Extract tax_code to find similar companies using HNSW

### Define function to query tax code

```python
def extract_tax_code(inp_tax_code, k):
  if inp_tax_code in tax_com_dict.keys():
    q_comp = tax_com_dict[inp_tax_code]
    print("Query company: ", q_comp, "\n\tTAX CODE: ", inp_tax_code)
    print()
    index = com_index_dictionary[q_comp]

    q_lab, q_dis = search(index, k)
    print("RECOMMENDED COMPANIES")
    for i in zip(q_lab[0], q_dis[0]):
      res_comp = ent_lis[i[0]]
      if res_comp != q_comp:
        print("-",res_comp, " \n\tDISTANCE: ", i[1], " \n\tTAX CODE: ", com_tax_dict[res_comp])
        print()
  else:
    print("Cannot find company")
```

### Call function to query tax code

```python
start_HNSW_Extract = time.time()

extract_tax_code('0101925883', 10)

end_HNSW_Extract = time.time()
print("\nTime taken to search using HNSW (ms): ", 1000*(end_HNSW_Extract - start_HNSW_Extract))
```

## Extract tax_code to find similar companies using scaNN

### Define function to query tax_code

```python
def extract_tax_code(inp_tax_code, k):
  if inp_tax_code in tax_com_dict.keys():
    q_comp = tax_com_dict[inp_tax_code]
    print("Query company: ", q_comp, "\n\tTAX CODE: ", inp_tax_code)
    print()
    index = com_index_dictionary[q_comp]

    query = np.array(com_emb.iloc[index])
    neighbors, distances = searcher.search(query, final_num_neighbors=k)
    print("RECOMMENDED COMPANIES")
    for i in zip(neighbors, distances):
      res_comp = ent_lis[i[0]]
      if res_comp != q_comp:
        print("-",res_comp, " \n\tDISTANCE: ", i[1], " \n\tTAX CODE: ", com_tax_dict[res_comp])
        print()
  else:
    print("Cannot find company")
```

### Call function to query tax_code

```python
start_scaNN_Extract = time.time()

extract_tax_code('0101925883',10)

end_scaNN_Extract  = time.time()
print("\nTime taken to search using scaNN (ms): ", 1000*(end_scaNN_Extract  - start_scaNN_Extract))
```
