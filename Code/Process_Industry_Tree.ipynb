# Create Industry Tree

## Create Entire Industry Tree

```python
indus_tree = {}
```

```python
for i in indus_name_lis:
  indus_tree[i] = {
      'parent': [],
      'children': [],
      'industry_number': []
  }
```

```python
for i in tqdm(indus_dict.keys()):
  for j in indus_dict[i]:
    if j[0] == '-1' and j[1] == '-1' and j[2] == '-1' and j[3] == '-1':
      indus_tree[j[4]]['industry_number'].append(i)
    elif j[0] != '-1':
      indus_tree[j[4]]['industry_number'].append(j[0])
      if (indus_num_to_indus_name[i] not in indus_tree[j[4]]['parent']):
        indus_tree[j[4]]['parent'].append(indus_num_to_indus_name[i])
      if (j[4] not in indus_tree[indus_num_to_indus_name[i]]['children']):
        indus_tree[indus_num_to_indus_name[i]]['children'].append(j[4])
      if j[1] != '-1':
        indus_tree[j[4]]['industry_number'].append(j[1])
        #indus_tree[j[4]]['parent'].append(indus_num_to_indus_name[i])
        #indus_tree[indus_num_to_indus_name[i]]['children'].append(j[4])
      if j[2] != '-1':
        indus_tree[j[4]]['industry_number'].append(j[2])
        #indus_tree[j[4]]['parent'].append(indus_num_to_indus_name[i])
        #indus_tree[indus_num_to_indus_name[i]]['children'].append(j[4])
      if j[3] != '-1':
        indus_tree[j[4]]['industry_number'].append(j[3])
        #indus_tree[j[4]]['parent'].append(indus_num_to_indus_name[i])
        #indus_tree[indus_num_to_indus_name[i]]['children'].append(j[4])
    elif j[0] == '-1':
      for k in range(1,4):
        if (j[k] != '-1'):
          indus_tree[j[4]]['industry_number'].append(j[k])
          prev = j[k][:-1]

          #Find parent
          while (len(prev) > 1 and indus_num_to_indus_name[prev] == j[4]):
            prev = prev[:-1]

          if (len(prev) > 1 and indus_num_to_indus_name[prev] != j[4]):
            if (indus_num_to_indus_name[prev] not in indus_tree[j[4]]['parent']):
              indus_tree[j[4]]['parent'].append(indus_num_to_indus_name[prev])
            if (j[4] not in indus_tree[indus_num_to_indus_name[prev]]['children']):
              indus_tree[indus_num_to_indus_name[prev]]['children'].append(j[4])

          #Find all parents
          while (len(prev) > 2 and indus_num_to_indus_name[prev[:-1]] == indus_num_to_indus_name[prev]):
            prev = prev[:-1]
            if (indus_num_to_indus_name[prev] not in indus_tree[j[4]]['parent']):
              indus_tree[j[4]]['parent'].append(indus_num_to_indus_name[prev])
            if (j[4] not in indus_tree[indus_num_to_indus_name[prev]]['children']):
              indus_tree[indus_num_to_indus_name[prev]]['children'].append(j[4])
```

### Define methods for Floyd Warshall Algorithm

```python
def floydWarshall(g):
    V = len(g)
    for k in tqdm(range(V)):
        for i in range(V):
            for j in range(V):
                if i != j:
                    g[i][j] = min(g[i][j], g[i][k] + g[k][j])
    return g
```
